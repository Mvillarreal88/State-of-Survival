```python
#importing 
import pandas as pd
```

```python
data_csv_path = "Resources/Life Expectancy.csv"
initial_data = pd.read_csv(data_csv_path)
print(len(initial_data))
initial_data.head()
```

```python
initial_data = initial_data.drop([327])
print(len(initial_data))
```

```python
initial_data["State"].unique()
```

```python
US_data = initial_data.iloc[:1]
US_data = US_data.drop(['County'], axis = 1)
US_data.head()
```

```python
state_county_df = initial_data.drop(labels = 0,
                                   axis = 0,
                                   inplace = False)
state_county_df.head()
```

```python
#Exporting US dataframe and county's Dataframe to CSV
state_county_df.to_csv("Data_csv/state_county.csv")
US_data.to_csv("Data_csv/US.csv")
```

```python
States_group = state_county_df.groupby(["County"])
States_group.head()
```

```python
#Creating a dataframe for the average life expectancy by State
state_data_df = state_county_df.loc[state_county_df['County'].isna()]
del state_data_df['County']
state_data_df = state_data_df.reset_index()
del state_data_df['index']
state_data_df
```

```python
state_data_df.to_csv("Data_csv/state_life_avg.csv")
```

```python
#Seperating the Data by regions
Northeast_life_expect = state_county_df.loc[state_county_df['State'] == ['Maine', 'Massachusetts', 'Rhode Island', 'Connecticut', 'New Hampshire', 'Vermont','New York', 'Pennsylvania', 'New Jersey', 'Delaware', 'Maryland']]
Northeast_life_expect
```
