```python
#importing 
import pandas as pd
import matplotlib.pyplot as plt
import scipy.stats as st
import numpy as np
from scipy.stats import linregress
```

```python
data_csv_path = "Resources/Obesity.csv"
initial_data = pd.read_csv(data_csv_path)
print(len(initial_data))
del initial_data['Unnamed: 10']
del initial_data['*Methodological changes in the 2011 BRFSS may affect comparability between the 2011 estimates and estimates from earlier years']
initial_data.head()
```

```python
#Dropping District of Colombia
initial_data = initial_data.drop([327])
initial_data["State"].unique()
```

```python
US_obes_data = initial_data.iloc[:1]
US_obes_data = US_obes_data.drop(['County'], axis = 1)
US_obes_data.head()
```

```python
state_county_obes = initial_data.drop(labels = 0,
                                   axis = 0,
                                   inplace = False)
state_county_obes.head()
```

```python
#Exporting US dataframe and county's Dataframe to CSV
state_county_obes.to_csv("Data_csv/state_county_obes.csv")
US_obes_data.to_csv("Data_csv/US_obes.csv")
```

```python
#Creating a dataframe for the Obesity by State
state_obes_df = state_county_obes.loc[state_county_obes['County'].isna()]
state_obes_df = state_obes_df.reset_index()
del state_obes_df['County']
del state_obes_df['index']
state_obes_df
```

```python
state_obes_df.to_csv("Data_csv/state_obes_avg.csv")
```

```python
#Seperating the Data by regions
Northeast_obes = state_obes_df[state_obes_df["State"].isin(['Maine', 'Massachusetts', 'Rhode Island','Connecticut',
                                                                 'New Hampshire', 'Vermont', 'New York', 
                                                                 'Pennsylvania', 'New Jersey', 'Delaware', 'Maryland'])]

Southeast_obes = state_obes_df[state_obes_df["State"].isin(['West Virginia', 'Virginia', 'Kentucky', 'Tennessee',
                                                                   'North Carolina', 'South Carolina', 'Georgia', 'Alabama',
                                                                   'Mississippi', 'Arkansas', 'Louisiana', 'Florida'])]
Midwest_obes = state_obes_df[state_obes_df["State"].isin(['Ohio', 'Indiana', 'Michigan', 'Illinois',
                                                                 'Missouri', 'Wisconsin', 'Minnesota', 'Iowa',
                                                                 'Kansas', 'Nebraska', 'South Dakota', 'North Dakota'])]
Southwest_obes = state_obes_df[state_obes_df["State"].isin(['Texas', 'Oklahoma', 'New Mexico', 'Arizona'])]

West_obes = state_obes_df[state_obes_df["State"].isin(['Colorado', 'Wyoming', 'Montana', 'Idaho',
                                                              'Washington', 'Oregon', 'Utah', 'Nevada',
                                                              'California', 'Alaska', 'Hawaii'])]
```

```python
#Exporting regional Data
Northeast_obes.to_csv("Data_csv/Northeast_obes.csv")
Southeast_obes.to_csv("Data_csv/Southeast_obes.csv")
Midwest_obes.to_csv("Data_csv/Midwest_obes.csv")
Southwest_obes.to_csv("Data_csv/Southwest_obes.csv")
West_obes.to_csv("Data_csv/West_obes.csv")
```
