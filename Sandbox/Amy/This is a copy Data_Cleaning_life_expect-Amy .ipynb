```python
#importing 
import pandas as pd
import matplotlib.pyplot as plt
import scipy.stats as st
import numpy as np
from scipy.stats import linregress
```

```python
data_csv_path = "Resources/Life Expectancy.csv"
initial_data = pd.read_csv(data_csv_path)
print(len(initial_data))
initial_data.head()
```

```python
#Dropping District of Colombia
initial_data = initial_data.drop([327])
print(len(initial_data))
```

```python
initial_data["State"].unique()
```

```python
US_data = initial_data.iloc[:1]
US_data = US_data.drop(['County'], axis = 1)
US_data.head()
```

```python
# DATA 
Life_Expectancy_by_Gender = ["Male life expectancy, 1985 (years)", "Male life expectancy, 2010 (years)", "Female life expectancy, 1985 (years)", "Female life expectancy, 2010 (years)"]
Years = [70.8, 76.1, 77.8, 80.8] 

x_axis = np.arange(0, len(Life_Expectancy_by_Gender))
tick_locations = []
for x in x_axis:
    tick_locations.append(x)
    

plt.title("Life Expectancy Change 1985 to 2010")
plt.xlabel("Gender")
plt.ylabel("Years")


plt.xlim(-0.5, len(Life_Expectancy_by_Gender)-.5)
plt.ylim(0, max(Years)+10)

plt.figure(figsize=(2,5))
plt.bar(x_axis, Years, facecolor="green", alpha=0.5, align="center")
plt.xticks(tick_locations, Life_Expectancy_by_Gender)

plt.show()
```

```python
state_county_df = initial_data.drop(labels = 0,
                                   axis = 0,
                                   inplace = False)
state_county_df.head()
```

```python
#Exporting US dataframe and county's Dataframe to CSV
state_county_df.to_csv("Data_csv/state_county.csv")
US_data.to_csv("Data_csv/US.csv")
```

```python
States_group = state_county_df.groupby(["County"])
States_group.head()
```

```python
#Creating a dataframe for the average life expectancy by State
state_data_df = state_county_df.loc[state_county_df['County'].isna()]
del state_data_df['County']
state_data_df = state_data_df.reset_index()
del state_data_df['index']
state_data_df
```

```python
state_data_df.to_csv("Data_csv/state_life_avg.csv")
```

```python
#Seperating the Data by regions
Northeast_life_expect = state_data_df[state_data_df["State"].isin(['Maine', 'Massachusetts', 'Rhode Island','Connecticut',
                                                                 'New Hampshire', 'Vermont', 'New York', 
                                                                 'Pennsylvania', 'New Jersey', 'Delaware', 'Maryland'])]

Southeast_life_expect = state_data_df[state_data_df["State"].isin(['West Virginia', 'Virginia', 'Kentucky', 'Tennessee',
                                                                   'North Carolina', 'South Carolina', 'Georgia', 'Alabama',
                                                                   'Mississippi', 'Arkansas', 'Louisiana', 'Florida'])]
Midwest_life_expect = state_data_df[state_data_df["State"].isin(['Ohio', 'Indiana', 'Michigan', 'Illinois',
                                                                 'Missouri', 'Wisconsin', 'Minnesota', 'Iowa',
                                                                 'Kansas', 'Nebraska', 'South Dakota', 'North Dakota'])]
Southwest_life_expect = state_data_df[state_data_df["State"].isin(['Texas', 'Oklahoma', 'New Mexico', 'Arizona'])]

West_life_expect = state_data_df[state_data_df["State"].isin(['Colorado', 'Wyoming', 'Montana', 'Idaho',
                                                              'Washington', 'Oregon', 'Utah', 'Nevada',
                                                              'California', 'Alaska', 'Hawaii'])]
```

```python
#Exporting regional Data
Northeast_life_expect.to_csv("Data_csv/Northeast_life_expect.csv")
Southeast_life_expect.to_csv("Data_csv/Southeast_life_expect.csv")
Midwest_life_expect.to_csv("Data_csv/Midwest_life_expect.csv")
Southwest_life_expect.to_csv("Data_csv/Southwest_life_expect.csv")
West_life_expect.to_csv("Data_csv/West_life_expect.csv")
```
